```python
#Đánh dấu các bản ghi mà Income null nhằm mục đích check lại sau này
import numpy as np
def TickNull(x):
    if np.isnan(x): # hoặc sử dụng kiểm tra bằng None
        return 'x'
    else:
        return ''

#Tạo 1 cột mới
#Với x là từng hàng trong data frame này, truyền giá trị tại cột E của X vào hàm để lấy về
dataset['IncomeIsNull'] = dataset.apply(lambda x: TickNull(x['Income']), axis=1)
dataset
```

```python
#Loại Outlier bằng phương pháp sử dụng IQR
def remove_outliers(df,colname):
    meangrade = df[colname].mean()
    q1 = df[colname].quantile(.25)
    q3 = df[colname].quantile(.75)
    iqr = q3 - q1
    toprange = q3 + iqr*1.5
    botrange = q1 - iqr*1.5
    df_result = df.drop(df[df[colname]>toprange].index)
    return df_result
```

```python
dfx_or = dfx #dfx Outliers Removed
dfx_or
```

```python
colname = ['Kidhome','Teenhome','Year_Birth','Income','NumWebVisitsMonth']
for i in colname:
    dfx_or = remove_outliers(dfx_or,i)
dfx_or
```
