```python
#Loại Outlier bằng phương pháp sử dụng IQR
def remove_outliers(df,colname):
    meangrade = df[colname].mean()
    q1 = df[colname].quantile(.25)
    q3 = df[colname].quantile(.75)
    iqr = q3 - q1
    toprange = q3 + iqr*1.5
    botrange = q1 - iqr*1.5
    df_result = df.drop(df[df[colname]>toprange].index)
    return df_result
```

3.2. PHÂN CỤM VỚI K-MEAN CLUSTERING

```python
df_clstr = dfx_or.copy()
df_clstr
```

```python
df_clstr.info()
```

```python
#Chuẩn hoá dữ liệu
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_clstr)

pd.DataFrame(data_scaled).describe()
```

```python
from sklearn.cluster import KMeans
wcss = []
for cluster in range(1,11):
    km = KMeans(n_clusters = cluster, init='k-means++')
    km.fit(data_scaled)
    wcss.append(km.inertia_)
print(wcss)
```

```python
#Vẽ biểu đồ
frame = pd.DataFrame({'Cluster':range(1,11), 'SSE':wcss})
plt.figure(figsize = (14,6))
plt.plot(frame['Cluster'], frame['SSE'], marker = 'o')
```

```python
#Chọn K = 4
```

```python
K = 4
Centroids = (df_clstr.sample(n = K))
color = ["gold","forestgreen","red","navy"]
plt.scatter(df_clstr["Income"],df_clstr["MntTotal"], c = 'darkgrey')

#Vẽ thêm tập tâm ngẫu nhiên
plt.scatter(Centroids["Income"],Centroids["MntTotal"], c = color)
plt.xlabel('Income')
plt.ylabel('MntTotal')
plt.show()
```

```python
import numpy as np
X = df_clstr[["Income","MntTotal"]]

diff = 1
j = 0
while (diff!=0):
    XD = X
    i = 1
    for index1, row_c in Centroids.iterrows():
        ED = []
        for index2, row_d in XD.iterrows():
            d1 = (row_c["Income"] - row_d["Income"])**2 #bình phương
            d2 = (row_c["MntTotal"] - row_d["MntTotal"])**2 #bình phương
            d = np.sqrt(d1+d2) #căn bậc hai của hai khoảng cách
            ED.append(d)
        X[i] = ED
        i = i+1
    C = []
    for index, row in X.iterrows():
        min_dist = row[1]
        pos = 1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos = i+1
        C.append(pos)
    X["Cluster"] = C
    Centroids_new = X.groupby("Cluster").mean()[["Income","MntTotal"]]
    if j ==0:
        diff=1
        j = j+1
    else:
        diff = (Centroids_new['MntTotal'] - Centroids['MntTotal']).sum() + (Centroids_new['Income'] - Centroids['Income']).sum()

    Centroids = X.groupby(["Cluster"]).mean()[["Income","MntTotal"]]
Centroids
```
